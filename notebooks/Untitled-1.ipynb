```python
# Correzione della classe Studente (indentazione e Media che restituisce un numero)
class Studente:
    def __init__(self, nome, cognome, corso_di_studi, listavoti):
        self.nome = nome
        self.cognome = cognome
        self.corso_di_studi = corso_di_studi
        self.listavoti = listavoti

    def Dati(self):
        return (
            f"Nome: {self.nome}\n"
            f"Cognome: {self.cognome}\n"
            f"Corso di studi: {self.corso_di_studi}\n"
            f"Lista voti: {self.listavoti}"
        )

    def Media(self):
        return sum(self.listavoti) / len(self.listavoti) if self.listavoti else 0.0

# Esempio di utilizzo
studente1 = Studente('Mario', 'Rossi', 'Ingegneria', [30, 18, 23])
studente2 = Studente('Pippo', 'Baudo', 'ScienzePol', [19, 18, 21])

print(studente1)
print(studente2)
print()
print(studente1.Dati())
print(f"Media voti: {studente1.Media():.2f}")
print()
print(studente2.Dati())
print(f"Media voti: {studente2.Media():.2f}")

print(type(studente1))
print(type(studente1.Media()))
print(Studente.Dati(studente1))
```
